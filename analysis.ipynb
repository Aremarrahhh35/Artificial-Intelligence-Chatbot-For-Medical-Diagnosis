```python
xpos = np.arange(len(analysis.prognosis))
xpos
```

```python
plt.xticks(xpos,analysis.prognosis)
plt.bar(xpos,analysis.chills)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Prognois')
plt.ylabel('Chills Symptom Frequency')
plt.title('Chills contribution to Illnesses')
```

```python
plt.yticks(xpos,analysis.prognosis)
plt.barh(xpos,analysis.chills)

plt.ylabel('Prognois')
plt.xlabel('Chills Symptom Frequency')
plt.title('Chills contribution to the Illnesses')
```

```python
fig= plt.figure(figsize=(10,8))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.barh(xpos,analysis.chills)
plt.yticks(xpos,analysis.prognosis)
plt.ylabel('Prognois')
plt.xlabel('Chills Symptom Frequency')
plt.title('Chills contribution to the Illnesses')
```

```python
fig= plt.figure(figsize=(10,8))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(xpos-0.2,analysis.chills)
axes.plot(xpos,analysis.fatigue)
axes.plot(xpos+0.2,analysis.body_pain)
plt.xticks(xpos,analysis.prognosis)
plt.xlabel('Prognois')
plt.ylabel('Chills Symptom Frequency')
plt.title('Chills contribution to the Illnesses')
```
